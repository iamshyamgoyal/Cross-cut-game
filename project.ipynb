```python
def winner_announcement(b):
    if winner_decision(b)==0:
        print('player 1 won')
    elif winner_decision(b)==1:
        print('player 2 won')
    else:
        print('take a new move')
```

```python
def possible_moves_2_P2(b):
  winning_move=[]
  optimal_move=[]
  all_possible_moves=[]
  for i in range(9):
    if b[i]==0:
      c=b.copy()
      c[i]=2
      if winner_decision(c)==1:
        winning_move.append(i+1)
      elif winner_decision(c)!=1:
        all_possible_moves.append(i+1)
      c[i]=1
      if winner_decision(c)==0:
        optimal_move.append(i+1)    
  if winning_move!=[]:
    print(f'winning move',winning_move)
    return winning_move
  elif optimal_move!=[]:
    print('optimal move',optimal_move)
    return optimal_move
  else:
    print(f'possible_moves {all_possible_moves}')
    return all_possible_moves   


      

```

```python
def possible_moves_P1(b):
    move_dictionary={1:[2,4,5],2:[1,3,5],3:[2,5,6],4:[1,5,7],5:[1,2,3,4,6,7,8,9],6:[3,5,9],7:[4,5,8],8:[5,7,9],9:[5,6,8]}
    possible_move_dict={}
    for i in move_dictionary:
        if b[i-1]==1:
            c=[]
            for k in move_dictionary[i]:
                if b[k-1]==0:
                    c.append(k)
            if c!=[]:
              possible_move_dict[i]=c
        else:
            None
       
            
    print(f'possible moves for player 1:{possible_move_dict}')               
    return possible_move_dict               
```
